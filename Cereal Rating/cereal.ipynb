```python
first=thinkstats2.Pmf((cereal.calories),label='calories' )
second=thinkstats2.Pmf((cereal.sodium),label='sodium')
width=5
axis=[0,310,0,0.5]
thinkplot.preplot(2,cols=2)
thinkplot.Hist(first,align='left',width=width)
thinkplot.Hist(second,align='right',width=width)
thinkplot.config(xlabel='sodium and calories',ylabel='pmf',axis=axis)
```
