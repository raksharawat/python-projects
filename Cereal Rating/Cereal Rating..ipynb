# Moments


# Find the first 3 raw moments of the amount of potassium present in the cereals

```python
def RawMoment(xs, k):
    return sum(x**k for x in xs) / len(xs)

RawMoment(cereal.potass,1), RawMoment(cereal.potass, 2), RawMoment(cereal.potass, 3)
```

# Find the first 3 central moments of the amount of fiber present in the cereals

```python
def CentralMoment(xs, k):
    mean = RawMoment(xs, 1)
    return sum((x - mean)**k for x in xs) / len(xs)

CentralMoment(cereal.fiber, 1), CentralMoment(cereal.fiber, 2), CentralMoment(cereal.fiber, 3)
```

# Find the moment about the mean  for the carbohydrates present in the cereal

```python
def Var(xs):
    return CentralMoment(xs, 2)

Var(cereal.carbo)
```

# Find the pearson median skewness of carbohydrates present in the cereals

```python
def Mean(xs):
    return RawMoment(xs, 1)

Mean(cereal.carbo)
```

```python
def Median(xs):
    cdf = thinkstats2.Cdf(xs)
    return cdf.Value(0.5)

Mean(cereal.carbo), Median(cereal.carbo)
```

# Find co-variance coefficient between fat present in cereal and protein present in cereal

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
Cov(cereal.fat,cereal.protein)
```

# Find correlation coefficient between fat present in cereal and protein present in cereal

```python
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
Corr(cereal.fat,cereal.protein)
```

# CONFIDENCE INTERVAL

# A sample of 6 cereals were taken and it was observed that average sodium content present in cereals was 154.16 mg and sample standard deviation was 82.60 . Find the 95 % confidence interval for the sodium content present in  cereals.

```python
c_s=pd.DataFrame(cereal.sodium)
c_s
```

```python
sod=c_s.head(6)
sod
```

```python
means=sod.mean()
std_dev=sod.std()
means,std_dev
```

```python
from scipy import stats
standard_error=stats.sem(sod)
conf_int=stats.t.interval(0.95,len(sod)-1,loc=means,scale=standard_error)
conf_int
```

# HYPOTHESIS TESTING
 

# When a sample of 45 cereals were taken a recent report claims that the amount of sugar present in cereals is less than 10 . If the average sugar content is 7.77 and population standard deviation is 4.44.Is the claim true ?Take alpha equal to 0.05

```python
c_s=pd.DataFrame(cereal.sugars)
c_s
```

```python
sug=c_s.head(45)
sug
```

```python
means=sug.mean()
std_dev=c_s.std()
means,std_dev
```

```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(c_s, x2=None, value=7.77)
print(float(pval))
if pval<0.05:
    print("Therefore we are reject null hypothesis")
else:
    print("Therefore we are not rejecting null hypothesis")
```

# CHI2

# A study was conducted and collected data on cereals. The data was classified into cereal type and Manufacturer. The data is as follows . Assume 95% confidence level

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

table = [[12,108],
         [24,156]]
print(table)
stat, p, dof, expected = chi2_contingency(table)
print('dof=%d' % dof)
print(expected)
prob = 0.95
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Therefore-we will Reject null hypothesis(H0)')
else:
    print('Therefore-Do not reject null hypothesis(H0)')
```
