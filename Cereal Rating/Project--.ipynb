# PMF

# SHOW THE DISTRIBUTION OF FAT IN BREAKFAST CEREALS

```python
hist = thinkstats2.Hist((cereal.fat), label='Fat present in Cereal')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Amount of fat', ylabel='Count of cereals')
```

```python
pmf = thinkstats2.Pmf(hist)
pmf
```

```python
thinkplot.Hist(pmf)
thinkplot.Config(xlabel='Amount of fat', ylabel='PMF')
```

```python
plt.figure(figsize=(16,9))
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel='Amount of fat', ylabel='Pmf')
```

# FIND THE AMOUNT OF PROTEIN PRESENT IN THE CEREALS AND SHOW IT GRAPHICALLY

```python
hist = thinkstats2.Hist((cereal.protein), label='Protein present in Cereal')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Amount of protein', ylabel='Count of cereals')
```

```python
pmf = thinkstats2.Pmf(hist)
pmf
```

```python
thinkplot.Hist(pmf)
thinkplot.Config(xlabel='Amount of protein', ylabel='PMF')
```

```python
pmf.Total()
```

# What is the probability that cereals of popular brand contain protein more than other brand?

```python
cdf = thinkstats2.Cdf((cereal.fiber), label='Fibre content in Cereal')
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Amount of fibre', ylabel='CDF')
```

```python
cdf = thinkstats2.Cdf(cereal.carbo, label='Carbohydrate content')
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Amount of Carbohydrates', ylabel='CDF', loc='upper left')
```

```python
cdf.Prob(12)
```

```python
plt.figure(figsize=(9,9))

cereal.sort_values('rating',ascending=False)
data1=cereal.head(5)

data2=cereal.tail(5)


first=thinkstats2.Cdf((data1.protein),label='Protein distribution for popular brand')
second=thinkstats2.Cdf((data2.protein),label='Protein distribution for other brand')


thinkplot.PrePlot(2)
thinkplot.Cdfs([first,second])
thinkplot.Config(xlabel='Amount of protein', ylabel='CDF')
```

# GEOMETRIC/BINOMIAL/POISSON DISTRIBUTION

```python
#GEOMETRIC DISTRIBUTION
geometric=np.random.geometric(p=0.8,size=int(1e6))
geo_pmf=thinkstats2.Pmf(geometric)
geo_pmf.Prob(4)
```

II. WHAT IS THE PROBABILITY THAT THAT THE CUSTOMER BOUGHT MORE THAN 4 TYPES OF CEREAL FROM THE GROCERY STORE ?

```python
#GEOMETRIC DISTRIBUTION
1-geo_pmf.Prob(4)
```
